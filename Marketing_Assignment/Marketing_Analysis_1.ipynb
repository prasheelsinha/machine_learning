Marketing mix stands as a widely utilized concept in the execution of marketing strategies. It encompasses various facets within a comprehensive marketing plan, with a central focus on the four Ps of marketing: product, price, place, and promotion

# Problem objective:
As a data scientist, you must conduct exploratory data analysis and hypothesis testing to enhance your comprehension of the diverse factors influencing customer acquisition.

# Data description:
The variables such as birth year, education, income, and others pertain to the first 'P' or 'People' in the tabular data presented to the user. The expenditures on items like wine, fruits, and gold, are associated with ‘Product’. Information relevant to sales channels, such as websites and stores, is connected to ‘Place’, and the fields discussing promotions and the outcomes of various campaigns are linked to ‘Promotion’

```python
# Since age has outliers, so using spearman correlation test to prove null hypothesis

# H0: With age, customers lean toward traditional in-store shopping preferences
# H1: With age, customers do not lean towards traditional in-store shopping preferences

from scipy.stats import spearmanr

corr, p_val = spearmanr(df.age, df.numstorepurchases)
print(f"Spearman correlation: {corr}, p-value: {p_val}")

if p_val < 0.05:
  print("Reject the null hypothesis and accept H1: With age, customers do not lean towards traditional in-store shopping preferences")
else:
  print("Failed to reject null hypothesis and accept H0: With age, customers lean toward traditional in-store shopping preferences")
```

```python
# H0: Customers with children likely experience time constraints, making online shopping a more convenient option.
# H1: Customer with children do not experience time constraints and do not lean towards online shopping.

import scipy.stats as stats

groups = [group.numwebpurchases.values for name, group in df.groupby('children')]

f_stat, p_val = stats.f_oneway(*groups)
print("F-statistic:", f_stat)
print("p-value:", p_val)

if p_val < 0.05:
  print("Reject the null hypothesis and accept H1: Customer with children do not experience time constraints and do not lean towards online shopping.")
else:
  print("Failed to reject null hypothesis and accept H0: Customers with children likely experience time constraints, making online shopping a more convenient option.")
```

```python
# Sales at physical stores may face the risk of cannibalization by alternative distribution channels.

dist_channels = ['numstorepurchases','numwebpurchases','numcatalogpurchases','numdealspurchases']

# Boxplot for above dist_channels
sns.boxplot(data=df[dist_channels])
plt.title('Boxplot of Distribution Channels')
plt.xlabel('Distribution Channels')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

sns.heatmap(df[dist_channels].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.xticks(rotation=45)

plt.show()
```

```python
# Using spearman hypothesis test because there are outliers in numdealspurchase column.

#H0: Sales at physical stores may face the risk of cannibalization by alternative distribution channels.
#H1: Sales at physical stores do not face the risk of cannibalization by alternative distribution channels.

from scipy.stats import spearmanr

dist_channels = ['numwebpurchases','numcatalogpurchases','numdealspurchases']

for channel in dist_channels:
  corr, p_val = spearmanr(df.numstorepurchases, df[channel])
  print(f"Spearman correlation for {channel}: {corr}, p-value: {p_val}")
  if p_val < 0.05:
    print(f"Reject the null hypothesis and accept H1: Sales at physical stores do not face the risk of cannibalization by {channel}")
  else:
    print(f"Failed to reject null hypothesis and accept H0: Sales at physical stores face the risk of cannibalization by {channel}")
  print(f"="*200)

```

```python
# Does the United States significantly outperform the rest of the world in total purchase volumes

# H0: US does not outperforms rest of the world in total purchases
# H1: US outperform rest of the world in total purchases

df.country.value_counts()

from scipy.stats import ttest_ind

us = df[df.country == 'US']['total_purchases']
rest = df[df.country != 'US']['total_purchases']

t_stat, p_val = ttest_ind(us, rest)
print("T-statistic:", t_stat)
print("p-value:", p_val)

if p_val < 0.05:
  print("Reject the null hypothesis and accept H1: US outperform rest of the world in total purchases")
else:
  print("Failed to reject null hypothesis and accept H0: US does not outperforms rest of the world in total purchases")
```

```python
# H0: There is no discernible pattern between children and total_purchases
# H1: There is a discernible pattern between children and total_purchases

# Using kruskal-wallis test as children column had outliers

from scipy.stats import kruskal

children_groups = [group.total_purchases.values for name, group in df.groupby('children')]

f_stat, p_val = stats.kruskal(*children_groups)
print("F-statistic:", f_stat)
print("p-value:", p_val)

if p_val<0.05:
  print("Reject the null hypothesis and accept H1: There is a discernible pattern between children and total_purchases")
else:
  print("Failed to reject null hypothesis and accept H0: There is no discernible pattern between children and total_purchases")
```
